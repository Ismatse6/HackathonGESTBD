```python
import os
import time
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from tqdm import tqdm

def DownloadPdf(url, folder):
    fileName = os.path.basename(url)
    destPath = os.path.join(folder, fileName)
    i = 1

    response = requests.get(url, headers=HEADERS, stream=True, verify=False)
    response.raise_for_status()

    with open(destPath, "wb") as f:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                f.write(chunk)

    return destPath


def ScrapeLinks(pageUrl, directory, headers):
    # Descargar la página y parsear HTML
    response = requests.get(pageUrl, headers=headers, verify=False)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")

    # Buscar todas las tablas
    tables = soup.find_all("table")

    # Extraer todos los enlaces dentro de las tablas
    pdfLinks = []
    for table in tables:
        for a in table.find_all("a", href=True):
            href = a["href"].strip()
            if href.lower().endswith(".pdf"):
                fullUrl = urljoin(pageUrl, href)
                pdfLinks.append(fullUrl)

    # Descargar todos los PDFs
    for pdfUrl in pdfLinks:
        try:
            downloadedPath = DownloadPdf(pdfUrl, directory)
            time.sleep(1)
        except Exception as e:
            print(f"ERROR al descargar {pdfUrl}: {e}")
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (compatible; PdfDownloader/1.0)"
}
DIRECTORY = "Guias Docentes"

os.makedirs(DIRECTORY, exist_ok=True)

for subject in tqdm(subjects, desc="Processing subjects"):
    ScrapeLinks(subject, DIRECTORY, HEADERS)
```
